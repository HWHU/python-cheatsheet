```python
import os

file_name_list = []

path = "C:/data/english"

# read filenames into a list

for file in os.listdir(path):
    file_name_list.append(file)
    
file_name_list
```

```python
import pandas as pd

# turn list into dataframe, add a column name
df = pd.DataFrame(file_name_list, columns=['filename'])
```

```python
df
```

```python
# add a column with name category
# get the numner of filenames in file_name_list
n = len(file_name_list)
# create n number of 'class1' label
category = ['class1']*n
# add category as a new column in df
df['category'] = category
```
