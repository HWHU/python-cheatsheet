```python
import os
import pandas as pd
```

```python
def filenamestodf(path, folder):
    file_name_list = []
    fullpath = path+folder
    
    # read filenames into a list
    for file in os.listdir(fullpath):
        folder_file = folder + file
        file_name_list.append(folder_file)

    #print(file_name_list)

    # turn list into dataframe, add a column name
    df = pd.DataFrame(file_name_list, columns=['filename'])
    l = len(file_name_list)
    return df, l
```

```python
# add a column with name category
def addLabel(l, label):

    # create n number of 'class1' label
    category = [label]*l
    # add category as a new column in df
    df['category'] = category
    return df
```
